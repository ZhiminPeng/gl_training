### Geographical Feature

Based on the description of the Kaggle competition, all the reported issues are from 4 cities. 
Oh! People from some cities might be more active on voting than others. But how can we know which city each issue belongs to.
CLUSTERING! Run kmeans on all the geographical coordinates!



```python
keylist = ['latitude', 'longitude']
lat_log_info = train_data.select_columns(keylist)

out = gl.kmeans.create(lat_log_info, num_clusters=4, max_iter=30)        

cluster_info = out.get('cluster_info')
center = gl.SFrame( {'lat': cluster_info['latitude'], 'log': cluster_info['longitude']})
center.save('./center.csv')

label = out.get('clusterid')
geo_label = gl.SArray(data = label['clusterid'], dtype=str)
print geo_label.head(5)
```

Let us first have a look at the correlation between geo_label and num_votes.

```python
import graphlab.aggregate as agg

def check_column_correlation(data, column_name, head=10):
    print data.groupby([column_name],
                       [agg.COUNT(), agg.AVG('num_votes'), 
                        agg.AVG('num_comments'), agg.AVG('num_views')]).head(head)
```

```python
check_column_correlation(train_data, 'geo_label_4', 4)
```

```python
train_data['summary_words'] = train_data['summary'].count_words()
train_data['description_words'] = train_data['description'].count_words()
```

##### Stop words

```python


stop_words = 'common-english-words.txt'
stop_words_list = []

for line in file('common-english-words.txt'):
    stop_words_list.extend(line.strip().split(','))

stop_word_set = set(stop_words_list)

def normalize_dict_values(d):
    total_value = np.sum(d.values())
    return dict([(k, float(v) / total_value) for k, v in d.iteritems()])
```

Remove stop words

```python
train_data['summary_words'] = train_data['summary_words'].apply(lambda x: dict([(k,v) for k,v in x.iteritems() 
                                                                                    if k not in stop_word_set])) 

train_data['description_words'] = train_data['description_words'].apply(lambda x: dict([(k,v) for k,v in x.iteritems() 
                                                                                    if k not in stop_word_set])) 
```

Let us look at the length of 'summary' and 'description'. 

```python
train_data['summary_length'] = train_data['summary_words'].apply(lambda x: len(x))
train_data['description_length'] = train_data['description_words'].apply(lambda x: len(x))
```

###### IDF -- Inversed Document Frequency
