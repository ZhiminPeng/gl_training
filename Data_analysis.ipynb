```python
import math
dword_dict = defaultdict(int)
for row in train_data['description_words']:
    for w, c in row.iteritems():
        dword_dict[w] += c

dword_dict = dict([(k, math.log(1. + v)) for k, v in dword_dict.iteritems()])
```

```python
new_desc_words =[]
for row in train_data['description_words']:
    new_row = dict()
    for k, v in row.iteritems():
        new_row[k] = v / (dword_dict[k])
    new_desc_words.append(new_row)
train_data['description_words'] = new_desc_words
```

```python
import numpy as np
def normalize_word_weights(data, column_name):
    new_col = []
    for row in data[column_name]:
        new_dict = {}
        row_sum = np.sum([v for k,v in row.iteritems()])
        new_dict = dict([(k, v / float(row_sum)) for k ,v in row.iteritems()])
        new_col.append(new_dict)
    return new_col
```

```python
train_data['summary_words'] = normalize_word_weights(train_data, 'summary_words')
```

```python
summary_dict = {}
description_dict = {}

for row in train_data['summary_words']:
    for k,v in row.iteritems():
        if summary_dict.has_key(k):
            continue
        else:
            summary_dict[k] = len(summary_dict)

for row in train_data['description_words']:
    for k,v in row.iteritems():
        if description_dict.has_key(k):
            continue
        else:
            description_dict[k] = len(description_dict) + len(summary_dict)
```

```python
print len(summary_dict), len(description_dict)
```

```python
from scipy.sparse import csr_matrix
```

```python
row_index  = []
col_index = []
val_index = []

cnt = 0
for row in  train_data['summary_words']:
    for k, v in row.iteritems():
        row_index.append(cnt)
        col_index.append(summary_dict[k])
        val_index.append(v)
    cnt += 1

cnt = 0

for row in  train_data['description_words']:
    for k, v in row.iteritems():
        row_index.append(cnt)
        col_index.append(description_dict[k])
        val_index.append(v)
    cnt += 1
    
```

```python
description_dict['street']
```

```python
data_matrix =  csr_matrix( (val_index,(row_index,col_index)), shape=(train_data.num_rows(), len(summary_dict) + len(description_dict)) )
```

```python
print gl.Sketch(train_data['tag_type'])
```

```python
tag_type_index = {}

for x in train_data['tag_type'].unique():
    if x == 'NA': continue
    tag_type_index[x] = len(tag_type_index)
```

```python
known_type_rows = []
unknown_type_rows = []

known_type_int = []

cnt = 0

for row in train_data['tag_type']:
    if row == 'NA':
        unknown_type_rows.append(cnt)
    else:
        known_type_rows.append(cnt)
        known_type_int.append(tag_type_index[row])
    cnt +=1 
```

```python
len(known_type_rows), len(unknown_type_rows), len(known_type_int)
```

```python
train_data_matrix = data_matrix[known_type_rows]
test_data_matrix = data_matrix[unknown_type_rows]
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import LinearSVC
```

```python
pred_type = OneVsRestClassifier(LinearSVC(random_state=0)).fit(train_data_matrix, known_type_int).predict(test_data_matrix)
```

```python
tag_type_inverse = dict([(v,k) for k,v in tag_type_index.iteritems()])
new_col = [None] * train_data.num_rows()
for k,v in  zip(unknown_type_rows, pred_type):
    new_col[k] = tag_type_inverse[v]
for k,v in zip(known_type_rows, known_type_int):
    new_col[k] = tag_type_inverse[v]
```
