```python
train_data['summary_words'] = train_data['summary'].count_words()
```

```python
train_data.head(1)
```

```python
from collections import defaultdict
word_dict = defaultdict(int)
for row in train_data['summary_words']:
    for w, c in row.iteritems():
        word_dict[w] += c
```

```python
word_count = [(k,v) for k, v in word_dict.iteritems()]
word_count = sorted(word_count, key=lambda x: x[1], reverse=True)
word_count[:10]
```

```python
train_data['description_words'] = train_data['description'].count_words()
```

```python
import math
dword_dict = defaultdict(int)
for row in train_data['description_words']:
    for w, c in row.iteritems():
        dword_dict[w] += c

dword_dict = dict([(k, math.log(1. + v)) for k, v in dword_dict.iteritems()])
```

```python
new_desc_words =[]
for row in train_data['description_words']:
    new_row = dict()
    for k, v in row.iteritems():
        new_row[k] = v / (dword_dict[k])
    new_desc_words.append(new_row)
train_data['description_words'] = new_desc_words
```

```python
train_data.head(5)
```

```python
import numpy as np
def normalize_word_weights(data, column_name):
    new_col = []
    for row in data[column_name]:
        new_dict = {}
        row_sum = np.sum([v for k,v in row.iteritems()])
        new_dict = dict([(k, v / float(row_sum)) for k ,v in row.iteritems()])
        new_col.append(new_dict)
    return new_col
```

```python
train_data['summary_words'] = normalize_word_weights(train_data, 'summary_words')
```

```python
train_data['description_words'] = normalize_word_weights(train_data, 'description_words')
```
